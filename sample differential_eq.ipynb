```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equation dy/dt = -y*t + 5
def function(y, t):
    dydt = -y * t + 5
    return dydt

# Initial condition
y0 = 1

# Time range
t = np.linspace(0, 5, 100)  

# Solve the ODE
y = odeint(function, y0, t)

# Plot the solution
plt.plot(t, y)
plt.xlabel("Time")
plt.ylabel("Y")
plt.title("Solving dy/dt ODE")
plt.grid()
plt.show()
```
