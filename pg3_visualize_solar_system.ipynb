```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Define data for planets
planet_names = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
semi_major_axis = [0.39, 0.72, 1.0, 1.52, 5.2, 9.58, 19.22, 30.05]  # in AU (astronomical units)
orbit_period = [0.24, 0.62, 1.0, 1.88, 11.86, 29.46, 84.01, 164.8]  # in y
plt.style.use("dark_background")
fig,ax=plt.subplots(figsize=(8,8))
ax.scatter(0,0,c="yellow",s=100,label="sun")
theta= np.linspace(0,2*np.pi , 100)
for i,planets in enumerate(planet_names):
    x=semi_major_axis[i]*np.cos(theta)
    y=semi_major_axis[i]*np.sin(theta)
    plt.plot(x,y,label=planets)
    plt.plot(semi_major_axis[i])
ax.set_xlabel('Distance (AU)')
ax.set_ylabel('Distance (AU)')
ax.set_title('Solar System')
ax.legend()

plt.grid(True)
plt.show()


```
