```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
m = 1000  # Mass of the rocket (kg)
thrust = 20000  # Thrust of the rocket (N)

def rocket_motion_ode(t, state):
    """
    Defines the differential equations for the rocket's motion.
    
    Parameters:
        t (float): Time variable.
        state (list): [position (x), velocity (v)].
    
    Returns:
        list: Derivatives [dx/dt = v, dv/dt = a].
    """
    x, v = state
    a = (thrust - m * g) / m  # Compute acceleration
    return [v, a]

# Initial conditions
x0 = 0  # Initial position (m)
v0 = 0  # Initial velocity (m/s)
state0 = [x0, v0]

# Time span for simulation
t_span = (0, 100)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Solve ODE
sol = solve_ivp(rocket_motion_ode, t_span, state0, t_eval=t_eval)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(sol.t, sol.y[0], label="Position (m)")
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Rocket Trajectory')
plt.legend()
plt.grid(True)
plt.show()
```
