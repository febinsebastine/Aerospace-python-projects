```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling import models
from astropy import units as u
from specutils.spectra import Spectrum1D
from specutils.fitting import fit_generic_continuum

# Step 1: Generate Random Spectra
np.random.seed(0)
x = np.linspace(0., 10., 200)
y = 3 * np.exp(-0.5 * (x - 6.3)**2 / 0.8**2)
y += np.random.normal(0., 0.2, x.shape)

spectra = Spectrum1D(flux=y*u.Jy, spectral_axis=x*u.um)
gaussian = models.Gaussian1D(amplitude=3.*u.Jy, mean=6.1*u.um, stddev=1.*u.um)
gaussian_fitted = fit_lines(spectra, gaussian)
y_gaussian_fitted = gaussian_fitted(x*u.um)

plt.plot(x, y, label="Original Spectrum", linestyle="dashed", color="blue")
plt.plot(x, y_gaussian_fitted, label="Fitted Gaussian", color="red")
plt.title("Fitting Spectra using Specutils")
plt.xlabel("Wavelength (m)")
plt.ylabel("Flux (Jy)")
plt.grid(True)
plt.legend()
plt.show()
```
