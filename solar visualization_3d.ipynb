```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define data for planets
planet_names = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
semi_major_axis = [0.39, 0.72, 1.0, 1.52, 5.2, 9.58, 19.22, 30.05]  # in AU (astronomical units)

# Set dark background style
plt.style.use('dark_background')

# Create figure and 3D axis
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the Sun at the center
ax.scatter(0, 0, 0, color='yellow', s=150, label='Sun')

# Plot orbits of planets
for i, planet in enumerate(planet_names):
    orbit_radius = semi_major_axis[i]
    orbit_theta = np.linspace(0, 2*np.pi, 200)  # Smooth circular path

    # Compute x, y coordinates
    x = orbit_radius * np.cos(orbit_theta)
    y = orbit_radius * np.sin(orbit_theta)
    z = np.zeros_like(orbit_theta)  # Flat orbits (z = 0)

    # Plot the 3D orbit
    ax.plot(x, y, z, label=planet)

# Set labels and title
ax.set_xlabel('Distance (AU)')
ax.set_ylabel('Distance (AU)')
ax.set_zlabel('Z-Axis (Flat)')
ax.set_title('3D Solar System Orbits (Without Inclination)')

# Equal aspect ratio


# Display legend
ax.legend()

# Show the plot
plt.show()
```
