```python
import json
import matplotlib.pyplot as plt

# Open the JSON file
with open('adsb_dataset.json', 'r') as f:
    data = json.load(f)['trace']

# Filter valid data
filtered_data = [entry for entry in data if entry[8] is not None]
airspeeds=[]
# Print first valid entry
print(filtered_data[0])
# Extract airspeed (True Air Speed - TAS)
for entry in filtered_data:
    if 'tas' in entry[8]:
        airspeeds.append([entry[8]['tas']])
# Plot the airspeed over time

plt.plot(airspeeds, color="red")
plt.xlabel("Data Point Index")
plt.ylabel("Airspeed (knots)")
plt.title("Aircraft Airspeed Over Time")
plt.grid(True)
plt.show()
```
